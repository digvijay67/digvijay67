```python
def check_data(df):
    print(80 * "*")
    print('DIMENSION: ({}, {})'.format(df.shape[0], df.shape[1]))
    print(80 * "*")
    print("COLUMNS:\n")
    print(df.columns.values)
    print(80 * "*")
    print("DATA INFO:\n")
    print(df.dtypes)
    print(80 * "*")
    print("MISSING VALUES:\n")
    print(df.isnull().sum())
    print(80 * "*")
    print("NUMBER OF UNIQUE VALUES:\n")
    print(df.nunique())
    
check_data(df)
```

```python
import statistics
import csv

with open('pima.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  
    data = [row for row in reader]

column1 = [float(row[0]) for row in data]
mean = statistics.mean(column1)
median = statistics.median(column1)
stdev = statistics.stdev(column1)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard deviation: {stdev}")
```
